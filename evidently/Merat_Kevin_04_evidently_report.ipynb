# Projet n°7 : Implémenter_un_modele_de_scoring

- Ce notebook est destiné à la création du tableau HTML d’analyse de data drift réalisé à partir de la librairie **evidently**.
- On prendra comme hypothèse que le dataset **application_train** représente les datas pour la modélisation et le dataset **application_test** représente les datas de nouveaux clients une fois le modèle en production.
-L’analyse du data drift permet d’identifier d’éventuelles évolutions dans les caractéristiques des clients (revenus, âge, situation professionnelle, etc...) susceptibles d’altérer les performances du modèle. Cette étape est essentielle dans un contexte de mise en production afin de mettre en place un monitoring des données et d’anticiper la nécessité d’un ré-entraînement du modèle.

```python
# Import des librairies nécessaires au projet. 

from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[0]
DATA_DIR = PROJECT_ROOT / "data" / "Donnee_kaggle"

import pandas as pd

from evidently import Report
from evidently.presets import DataDriftPreset
```

```python
# Import des différents fichiers du projet.

df_appli_train = pd.read_csv(DATA_DIR / "application_train.csv")
df_appli_test = pd.read_csv(DATA_DIR / "application_test.csv")
```

```python
# Mise en place du rapport evidently pour analyser le data drift.

# Suppression de la target et de l'identifiant
df_appli_train = df_appli_train.drop(columns=["TARGET", "SK_ID_CURR"])
df_appli_test = df_appli_test.drop(columns=["SK_ID_CURR"])

# Rapport Data Drift
report = Report([DataDriftPreset()])
eval_report = report.run(current_data=df_appli_test, reference_data=df_appli_train)

# Sauvegarder en HTML
output_html = "evidently_data_drift_report.html"
eval_report.save_html(output_html)

print("Rapport généré :", output_html)
```
